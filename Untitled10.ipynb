```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate example data with 8 time series variables
num_periods = 20
time_steps = 50
x = np.linspace(0, num_periods, time_steps)
y = np.linspace(0, 1, 8)
X, Y = np.meshgrid(x, y)
Z = np.random.rand(8, time_steps) * np.sin(X * np.pi / num_periods)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Time')
ax.set_ylabel('Variable')
ax.set_zlabel('Value')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='coolwarm')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some dummy data
interest_rates = np.random.rand(20, 240, 8)

# Set up the figure and axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Set up the X and Y axes
X = np.arange(240)
Y = np.arange(8)
X, Y = np.meshgrid(X, Y)

# Plot the surfaces
for bond in range(20):
    Z = interest_rates[bond, :, :].T
ax.plot_surface(X, Y, Z, cmap='coolwarm')

plt.show()
```
